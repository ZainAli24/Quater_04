### 3: URLs and Email Addresses:
- To quickly turn a URL or email address into a link, enclose it in angle brackets:
    - <https://github.com/ZainAli24?tab=repositories    >
    - <fake@example.com>


### 4. Formatting Links:
- To emphasize links, add asterisks before and after the brackets and parentheses. To denote links as code, add backticks in the brackets.
    - MY Favourite Repo is **[Zain GitHub Repo](https://github.com/ZainAli24?tab=repositories)**
    - MY Favourite Repo is *[Zain GitHub Repo](https://github.com/ZainAli24?tab=repositories)*
    - MY Favourite Code is [`Zain GitHub Repo Code`](https://github.com/ZainAli24?tab=repositories "SEE Code")
